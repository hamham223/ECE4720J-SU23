```python
import matplotlib.pyplot as plt

# Data for line chart
years = [2020, 2021, 2022]
gpa = [3.3, 3.72, 3.87]

# Data for quartile plot
ranks = [150/220, 37/219, 22/146]

# Create line chart
plt.plot(years, gpa, marker='o', label='GPA')

# Create quartile plot
plt.scatter(years, ranks, marker='x', color='r', label='Rank')

# Set plot labels and title
plt.xlabel('Years')
plt.ylabel('GPA / Rank')
plt.title('GPA and Rank Comparison')

# Set legend
plt.legend()

# Display the plot
plt.show()
```
